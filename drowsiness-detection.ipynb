```python
import os

from ultralytics import YOLO
```

```python
# Load a model
model = YOLO("yolov8n.pt")  # load pre trained model

# Use the model
results = model.train(data="/kaggle/input/drowsy/Drowsy Driver Detection.v8i.yolov8/data.yaml", epochs=22)  # train the model
```
